```python
from tkinter import *
import time
import math
tick = ""
reps = 0
counts = None
# ---------------------------- CONSTANTS ------------------------------- #
PINK = "#e2979c"
RED = "#e7305b"
GREEN = "#9bdeac"
YELLOW = "#f7f5dd"
FONT_NAME = "Courier"
WORK_MIN = 25
SHORT_BREAK_MIN = 5
LONG_BREAK_MIN = 20
BLUE = "#008DDA"

window = Tk()

def reset():
    global tick
    global reps
    window.after_cancel(counts)
    Timer_label.config(text="Timer", fg=GREEN, bg=YELLOW, font=(FONT_NAME,35,'bold'))
    canvas.itemconfig(timer, text="00:00")
    tick_label.config(text="")
    tick=""
    reps=0
def count(c):
    
    global reps
    global tick
    min = math.floor(c/60)
    sec = c % 60
    
    if len(str(sec)) < 2: 
        sec = "0"+str(sec)
        
    if len(str(min)) < 2: 
        min = "0"+str(min)
        
    if c > -1:
        global counts
        counts = window.after(1000, count, c-1)
        canvas.itemconfig(timer, text=f"{min}:{sec}")
    else:
        start()
        
    
        if reps in [3,5,7,9]:
            tick += "✔"
        tick_label.config(text=tick)
    
    
    
def start():
    global reps
    reps += 1
    
    work_sec = WORK_MIN * 60
    short_break_sec = SHORT_BREAK_MIN * 60
    long_break_sec = LONG_BREAK_MIN * 60
    
    if reps % 8 == 0:
        count(long_break_sec)
        Timer_label.config(text='Break', fg=RED)
        
    elif reps % 2 == 0:
        count(short_break_sec)
        Timer_label.config(text='Break', fg=PINK)
    else:
        count(work_sec)
        Timer_label.config(text='Work', fg=GREEN)
        

    
window.title("Pomodora")
window.config(padx=100, pady=50, bg=YELLOW, highlightbackground="#FFC700",highlightthickness=15, highlightcolor="#FFF455")

canvas = Canvas(width=200, height=224, bg=YELLOW, highlightthickness=0)
tomato_img = PhotoImage(file=r'C:\Users\nickf\OneDrive\桌面\Python\Git\Python-coding-skills\img_materials_timer\tomato.png')
canvas.create_image(100, 112, image=tomato_img)
timer = canvas.create_text(103,130, text="00:00",fill="white", font=(FONT_NAME, 35,"bold"))
canvas.grid(column=1,row=1)           
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
Timer_label = Label(text="Timer", fg=GREEN, bg=YELLOW, font=(FONT_NAME,35,'bold'))
Timer_label.grid(column=1,row=0)

Start_button = Button(text="Start",highlightthickness=0, command=start)
Start_button.grid(column=0, row=3)

Reset_button = Button(text='Reset',highlightthickness=0, command=reset)
Reset_button.grid(column=2, row=3)

tick_label = Label(text="", fg=GREEN, bg=YELLOW)
tick_label.grid(column=1, row=4)
window.mainloop()
```
