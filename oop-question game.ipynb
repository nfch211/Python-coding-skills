```python
import os
import sys
import pandas as pd
sys.path.append(r"C:\Users\nickf\OneDrive\桌面\Python")

from my_functions.functions import *
import ast
import json
path1 =r'C:\Users\nickf\OneDrive\桌面\day17\data.py'


lst = get_list(r'C:\Users\nickf\OneDrive\桌面\day17',prt=True)

sys.path.append(r'C:\Users\nickf\OneDrive\桌面\day17')
from data import question_data

class Question:
    def __init__(self, text, answer):
        self.text= text
        self.answer = answer
    
question_bank=[]    

for question in question_data:
    new_question = Question(question['text'], question['answer'])
    question_bank.append(new_question)

class QuizBrain:
    def __init__(self, question):
        self.question_number = 0
        self.question_list = question
        self.score = 0
    def next_question(self):
        current_question = self.question_list[self.question_number]
        self.question_number +=1
        user_answer = input(f"Q.{self.question_number}:{current_question.text} True or False? ")
        self.check_answer(user_answer, current_question.answer)
        
    def still_has_question(self):
        return len(self.question_list) > self.question_number
    
    def check_answer(self, user_answer, correct_answer):
        if user_answer.lower() == correct_answer.lower():
            self.score +=1
            print("You are correct!")
        
        else:
            print("You are wrong!")
        print(f"The correct answer is: {correct_answer}")
        total_score = f'{self.score}/{len(self.question_list)}'
        print(f"Your current score is: {total_score}")

quiz = QuizBrain(question_bank)



while quiz.still_has_question():
    quiz.next_question()

print(f"Your final score is: {quiz.score}/{quiz.question_number}")
```
